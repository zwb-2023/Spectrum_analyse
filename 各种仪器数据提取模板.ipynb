```python
import pandas as pd  # 导入pandas库，用于数据处理和分析
import numpy as np
import glob

list_path = glob.glob(rf'I:\experiment_data\莲蓉混合\纯蓉\*.csv')
df = pd.read_csv(list_path[0])  # 使用read_csv函数读取CSV文件
wavelenth = df.columns.tolist()[4:-1]
name = df.iloc[:,0].tolist()
df1 = df.iloc[:,4:-1]
df1.index = name
mean = np.mean(np.array(df1) , axis=0)

# 替换wavelenth里面所有字符串为float
for i in range(len(wavelenth)):
  wavelenth[i] = float(wavelenth[i])
wavelengths = 10000000/np.array(wavelenth)
```

```python
def get_mean(path):
  df = pd.read_csv(path)  # 使用read_csv函数读取CSV文件，并且重复数据取平均
  name = df.iloc[:,0].tolist()
  df1 = df.iloc[:,4:-1]
  df1.index = name
  mean = np.mean(np.array(df1) , axis=0)
  labels = name[0]
  return mean , labels
```

```python
def get_data(path):
  df = pd.read_csv(path)  # 使用read_csv函数读取CSV文件，并且重复数据取平均
  name = path.split('\\')[-1][:-4]
  df1 = df.iloc[:,4:-1]
  data = np.array(df1)
  labels = name
  return data , labels
```

```python
datas = [] 
labels = []
for path in list_path :
  data , label = get_data(path)
  datas.append(data[0])
  labels.append(label)

datas = np.array(datas)
labels = np.array(labels)
```

```python
df = pd.DataFrame(datas , index = labels , columns = wavelengths)
df
```

### 海谱香精检测仪数据提取

```python
import xml.etree.ElementTree as ET

def extract_wavelength(node):
    node_data = node[0]
    sample_nodes = node_data.findall('Sample')
    test_node = sample_nodes[0]
    polyline = test_node.find('polyline')
    
    value = polyline.get('points')
    value = str(value)
    str_list = value.split()
    k = len(str_list)
    
    wavelength = [None] * k
    for i in range(k):
        temp = str_list[i].split(",")
        wavelength[i] = temp[0]
    
    return wavelength

def extract_wavelength_for_all(node):
    test_node = node[0]
    polyline = test_node.find('polyline')
    
    value = polyline.get('points')
    value = str(value)
    str_list = value.split()
    k = len(str_list)
    
    wavelength = [None] * k
    for i in range(k):
        temp = str_list[i].split(",")
        wavelength[i] = temp[0]
    
    return wavelength

def read_dat_xml_fun(path):
    dom = ET.parse(path)

    ir_node = dom.findall('.//SWIRChart')
    vis_node = dom.findall('.//VISChart')
    uv_node = dom.findall('.//UVChart')

    ir_avg_node = dom.findall('.//SWIRAvgChart')
    vis_avg_node = dom.findall('.//VISAvgChart')
    uv_avg_node = dom.findall('.//UVAvgChart')
    all_avg_node = dom.findall('.//ALLAvgChart')

    ir_scatter_node = dom.findall('.//SWIRAvgScatter')
    vis_scatter_node = dom.findall('.//VISAvgScatter')
    uv_scatter_node = dom.findall('.//UVAvgScatter')
    all_scatter_node = dom.findall('.//ALLAvgScatter')

    table_node = dom.findall('.//Table')

    # 提取波长信息
    ir_wavelength = extract_wavelength(ir_node)
    uv_vis_wavelength = extract_wavelength(vis_node)
    all_wavelength = extract_wavelength_for_all(all_avg_node)

    # 提取原始数据
    ir_node_cell = extract_points(ir_node)
    vis_node_cell = extract_points(vis_node)
    uv_node_cell = extract_points(uv_node)

    # 提取透射平均后数据
    ir_avg_cell = extract_avg_points(ir_avg_node)
    vis_avg_cell = extract_avg_points(vis_avg_node)
    uv_avg_cell = extract_avg_points(uv_avg_node)
    all_avg_cell = extract_avg_points(all_avg_node)

    # 提取散点坐标数据
    ir_scatter_cell = extract_scatter_points(ir_scatter_node)
    vis_scatter_cell = extract_scatter_points(vis_scatter_node)
    uv_scatter_cell = extract_scatter_points(uv_scatter_node)
    all_scatter_cell = extract_scatter_points(all_scatter_node)

    # 提取一致性结果表格数据
    node_data = table_node[0]
    row_nodes = node_data.findall('Row')
    n = len(row_nodes)

    table_data = [[None] * 4 for _ in range(n)]
    order = [None] * n  # Table的sampleID顺序就是order
    for i, test_data in enumerate(row_nodes):
        cell_nodes = test_data.findall('Cell')
        
        k = len(cell_nodes)
        for j in range(1, k - 1):
            test_cell = cell_nodes[j]
            value = test_cell.text
            table_data[i][j - 1] = float(value)
        
        value = test_data.get('sampleID')
        order[i] = float(value)

    # 提取背景和source
    ir_bg_node = dom.findall('.//SWIRAvgBg')
    vis_bg_node = dom.findall('.//VISAvgBg')
    uv_bg_node = dom.findall('.//UVAvgBg')

    ir_src_node = dom.findall('.//SWIRAvgSrc')
    vis_src_node = dom.findall('.//VISAvgSrc')
    uv_src_node = dom.findall('.//UVAvgSrc')

    ir_bg_cell = extract_avg_points(ir_bg_node)
    vis_bg_cell = extract_avg_points(vis_bg_node)
    uv_bg_cell = extract_avg_points(uv_bg_node)

    ir_src_cell = extract_avg_points(ir_src_node)
    vis_src_cell = extract_avg_points(vis_src_node)
    uv_src_cell = extract_avg_points(uv_src_node)

    # 重排顺序
    real_order = [i[0] for i in sorted(enumerate(order), key=lambda x: x[1])]

    # 好几个数据都要重排，除了table和scatter
    ir_node_cell = [ir_node_cell[i] for i in real_order]
    vis_node_cell = [vis_node_cell[i] for i in real_order]
    uv_node_cell = [uv_node_cell[i] for i in real_order]

    ir_avg_cell = [ir_avg_cell[i] for i in real_order]
    vis_avg_cell = [vis_avg_cell[i] for i in real_order]
    uv_avg_cell = [uv_avg_cell[i] for i in real_order]
    all_avg_cell = [all_avg_cell[i] for i in real_order]

    # 提取名称
    name_cell = [cell_nodes[0].text for cell_nodes in row_nodes]
    name_cell = [name_cell[i] for i in real_order]

    # 统一到一个结构里面保存
    result = {
        'ir_wavelength': ir_wavelength,
        'uv_vis_wavelength': uv_vis_wavelength,
        'all_wavelength': all_wavelength,

        'ir_node_cell': ir_node_cell,
        'vis_node_cell': vis_node_cell,
        'uv_node_cell': uv_node_cell,

        'ir_avg_cell': ir_avg_cell,
        'vis_avg_cell': vis_avg_cell,
        'uv_avg_cell': uv_avg_cell,
        'all_avg_cell': all_avg_cell,

        'ir_scatter_cell': ir_scatter_cell,
        'vis_scatter_cell': vis_scatter_cell,
        'uv_scatter_cell': uv_scatter_cell,
        'all_scatter_cell': all_scatter_cell,

        'ir_bg_cell': ir_bg_cell,
        'vis_bg_cell': vis_bg_cell,
        'uv_bg_cell': uv_bg_cell,

        'ir_src_cell': ir_src_cell,
        'vis_src_cell': vis_src_cell,
        'uv_src_cell': uv_src_cell,

        'table_data': table_data,
        'name_cell': name_cell
    }

    return result

def extract_avg_points(node):
    node_data = node[0]
    sample_nodes = node_data.findall('polyline')
    n = len(sample_nodes)

    data = sample_nodes[0]
    value = data.get('points')
    value = str(value)
    str_list = value.split()
    k = len(str_list)

    sample_cell = [[None] * k for _ in range(n)]

    for i, data in enumerate(sample_nodes):
        value = data.get('points')
        value = str(value)
        str_list = value.split()

        for j, temp in enumerate(str_list):
            sample_cell[i][j] = float(temp.split(',')[1])
    
    return sample_cell


def extract_points(node):
    node_data = node[0]
    sample_nodes = node_data.findall('Sample')
    n = len(sample_nodes)
    sample_cell = [None] * n

    for i, data in enumerate(sample_nodes):
        polyline = data.findall('polyline')
        m = len(polyline)
        test = polyline[0]
        value = test.get('points')
        value = str(value)
        str_list = value.split()
        k = len(str_list)

        sample_data = [[None] * k for _ in range(m)]

        for j, test in enumerate(polyline):
            value = test.get('points')
            value = str(value)
            str_list = value.split()

            for w, temp in enumerate(str_list):
                sample_data[j][w] = float(temp.split(',')[1])

        sample_cell[i] = sample_data

    return sample_cell

def extract_scatter_points(node):
    node_data = node[0]
    sample_nodes = node_data.findall('polyline')
    n = len(sample_nodes)

    sample_cell = [None] * n

    for i, data in enumerate(sample_nodes):
        value = data.get('points')
        value = str(value)
        str_list = value.split()
        k = len(str_list)

        scatter_data = [[None] * 2 for _ in range(k)]

        for j, temp in enumerate(str_list):
            temp_split = temp.split(',')
            scatter_data[j] = [float(temp_split[0]), float(temp_split[1])]

        sample_cell[i] = scatter_data

    return sample_cell
```

```python
import glob
import numpy as np
import pandas as pd
file_list = glob.glob(rf'I:\experiment_data\洗脱液\不同倍数洗脱液\*.dat')

transmittance = []
labels = []
for file in file_list : 
    result = read_dat_xml_fun(file)
    transmittance.append(np.array(result['all_avg_cell']))
    labels.append([file.split('\\')[-1][:-4]]*np.array(result['all_avg_cell']).shape[0])
                                     
labels = np.array(labels)
labels = labels.reshape(-1)
transmittance = np.concat(transmittance)
wavelengths =[float(x) for x in result['all_wavelength']]
transmittance_pd = pd.DataFrame(transmittance, index=labels, columns=wavelengths)
# transmittance_pd = transmittance_pd.groupby(transmittance_pd.index).first()
transmittance_pd
```
